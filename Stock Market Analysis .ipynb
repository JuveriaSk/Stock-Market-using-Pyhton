```python
tick1='AMZN'
tick2='AAPL'
tick3='GOOG'
tick4='MSFT'

start = datetime(end.year-4,11,end.day-3)
end_d = datetime(end.year,11,end.day-3)


AMZN=yf.download(tick1,start,end_d)
AAPL=yf.download(tick2,start,end_d)
GOOG=yf.download(tick3,start,end_d)
MSFT=yf.download(tick4,start,end_d)
```

#### Add a new column ‘Trend’ whose values are based on the 'Daily Return'.


```python
import numpy as np

# Define the trend function
def trend(x):
  if x > -0.015 and x <= 0.015:
    return 'Slight or No change'
  elif x > 0.015 and x <= 0.04:
    return 'Slight Positive'
  elif x < -0.015 and x >= -0.4:
    return 'Slight Negative'
  elif x > 0.04 and x <= 0.06:
    return 'Positive'
  elif x < -0.04 and x >= -0.06:
    return 'Negative'
  elif x > 0.06 and x <= 0.07:
    return 'Among top gainers'
  elif x < -0.06 and x >= -0.07:
    return 'Among top losers'
  elif x > 0.07:
    return 'Bull run'
  elif x <= -0.07:
    return 'Bear drop'
AMZN['Trend']= np.zeros(AMZN['Daily Return'].count()+1)
AMZN['Trend']= AMZN['Daily Return'].apply(lambda x:trend(x))
AMZN.tail(5)
```

```python
GOOG['Trend']= np.zeros(AMZN['Daily Return'].count()+1)
GOOG['Trend']= AMZN['Daily Return'].apply(lambda x:trend(x))
GOOG.tail(5)
```

```python
AAPL['Trend']= np.zeros(AMZN['Daily Return'].count()+1)
AAPL['Trend']= AMZN['Daily Return'].apply(lambda x:trend(x))
AAPL.tail(5)
```

```python
MSFT['Trend']= np.zeros(AMZN['Daily Return'].count()+1)
MSFT['Trend']= AMZN['Daily Return'].apply(lambda x:trend(x))
MSFT.tail(5)
```

#### What was the correlation between the daily returns of different stocks?

```python
all_adj = pd.concat([AAPL['Adj Close'] ,GOOG['Adj Close'],AMZN['Adj Close'],MSFT['Adj Close']], axis = 1)
```

```python
all_adj.columns = ['Apple','Google','Amazon','Microsoft']
```

```python
pct_df = all_adj.pct_change()
pct_df.tail()
```

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(pct_df.corr(), annot=True, linewidths=0.5, fmt=".1f", ax=ax, cmap="crest")
```
